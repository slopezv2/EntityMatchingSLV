```python
import py_stringmatching as sm
```

```python
def calculate_levenshtain_sim(df: pd.DataFrame, col_name_1: str, col_name_2: str):
    lev = sm.Levenshtein()
    df.loc[:,"levenshtain_distance"] = df.apply(lambda row: 1 - lev.get_sim_score(row[col_name_1], row[col_name_2]),axis=1)
    return df
```

```python
def calculate_NeedlemanWunsch_sim(df: pd.DataFrame, col_name_1: str, col_name_2: str):
    needleman_wunsch = sm.NeedlemanWunsch( gap_cost=0.5, sim_func=lambda a, b: 1 if a==b else 0)
    df.loc[:,"needleman_wunsch_distance"] = df.apply(lambda row: 1 - (needleman_wunsch.get_raw_score(row[col_name_1], row[col_name_2])/max(len(row[col_name_1]),len(row[col_name_2]))),axis=1)
    return df
```

```python
def calculate_Affine_Gap_sim(df: pd.DataFrame, col_name_1: str, col_name_2: str):
    af_gap = sm.Affine(gap_start=0.5,gap_continuation=0.2,sim_func=lambda a, b: 1 if a==b else 0)
    df.loc[:,"affine_gap_distance"] = df.apply(lambda row: 1 - (af_gap.get_raw_score(row[col_name_1], row[col_name_2])/max(len(row[col_name_1]),len(row[col_name_2]))),axis=1)
    return df
```

```python
def calculate_Smith_Waterman_sim(df: pd.DataFrame, col_name_1: str, col_name_2: str):
    sw = sm.SmithWaterman(gap_cost=1,sim_func=lambda a, b: 1 if a==b else 0)
    df.loc[:,"smith_waterman_distance"] = df.apply(lambda row: 1 - (sw.get_raw_score(row[col_name_1], row[col_name_2])/max(len(row[col_name_1]),len(row[col_name_2]))),axis=1)
    return df
```

```python
def calculate_jaro_sim(df: pd.DataFrame, col_name_1: str, col_name_2: str):
    jaro = sm.Jaro()
    df.loc[:,"jaro_distance"] = df.apply(lambda row: 1 - (jaro.get_raw_score(row[col_name_1], row[col_name_2])/max(len(row[col_name_1]),len(row[col_name_2]))),axis=1)
    return df
```

```python
def calculate_Jaro_Winkler_sim(df: pd.DataFrame, col_name_1: str, col_name_2: str):
    jw = sm.JaroWinkler(prefix_weight=0.25)
    df.loc[:,"jaro_winkler_distance"] = df.apply(lambda row: 1 - (jw.get_sim_score(row[col_name_1], row[col_name_2])),axis=1)
    return df
```

```python
def calculate_Overlap_Coefficient_sim(df: pd.DataFrame, col_name_1: str, col_name_2: str):
    oc = sm.OverlapCoefficient()
    df.loc[:,"overlap_coefficient_distance"] = df.apply(lambda row: 1 - (oc.get_sim_score(row[col_name_1].split(" "), row[col_name_2].split(" "))),axis=1)
    return df
```

```python
def calculate_Generalized_Jaccard_sim(df: pd.DataFrame, col_name_1: str, col_name_2: str):
    gj = sm.GeneralizedJaccard(sim_func=lambda a, b: 1 if a==b else 0)
    df.loc[:,"generalized_jaccard_distance"] = df.apply(lambda row: 1 - (gj.get_sim_score(row[col_name_1].split(" "), row[col_name_2].split(" "))),axis=1)
    return df
```

```python
def calculate_tfidf_sim(df: pd.DataFrame, col_name_1: str, col_name_2: str):
    tfidf = sm.TfIdf(dampen=False)
    df.loc[:,"tfidf_distance"] = df.apply(lambda row: 1 - (tfidf.get_sim_score(row[col_name_1].split(" "), row[col_name_2].split(" "))),axis=1)
    return df
```

```python
def calculate_soft_tfidf_sim(df: pd.DataFrame, col_name_1: str, col_name_2: str):
    softtfidf = sm.SoftTfIdf()
    df.loc[:,"soft_tfidf_distance"] = df.apply(lambda row: 1 - (softtfidf.get_raw_score(row[col_name_1].split(" "), row[col_name_2].split(" "))/max(len(row[col_name_1]),len(row[col_name_2]))),axis=1)
    return df
```

```python
def calculate_partial_ration_sim(df: pd.DataFrame, col_name_1: str, col_name_2: str):
    pr = sm.PartialRatio()
    df.loc[:,"partial_ration_distance"] = df.apply(lambda row: 1 - (pr.get_sim_score(row[col_name_1], row[col_name_2])),axis=1)
    return df
```

```python
def calculate_bag_distance_sim(df: pd.DataFrame, col_name_1: str, col_name_2: str):
    bd = sm.BagDistance()
    df.loc[:,"bag_distance_distance"] = df.apply(lambda row: 1 - (bd.get_sim_score(row[col_name_1], row[col_name_2])),axis=1)
    return df
```

```python
str_path = "../../Datos/products_csv/df_price_cross_cpus.parquet"
```

```python
import pandas as pd
from sklearn.preprocessing import minmax_scale
```

```python
def process_features(df: pd.DataFrame,dest_path: str):
    calculate_levenshtain_sim(df,"ProductID","ProductID2")
    calculate_NeedlemanWunsch_sim(df,"ProductID","ProductID2")
    calculate_Affine_Gap_sim(df,"ProductID","ProductID2")
    calculate_Smith_Waterman_sim(df,"ProductID","ProductID2")
    calculate_jaro_sim(df,"ProductID","ProductID2")
    calculate_Jaro_Winkler_sim(df,"ProductID","ProductID2")
    calculate_Overlap_Coefficient_sim(df,"ProductID","ProductID2")
    calculate_Generalized_Jaccard_sim(df,"ProductID","ProductID2")
    calculate_tfidf_sim(df,"ProductID","ProductID2")
    calculate_soft_tfidf_sim(df,"ProductID","ProductID2")
    calculate_partial_ration_sim(df,"ProductID","ProductID2")
    calculate_bag_distance_sim(df,"ProductID","ProductID2")
    df[["same_product"]] = df[["same_product"]].astype(int)
    df[["levenshtain_distance", "needleman_wunsch_distance", "affine_gap_distance", 
        "smith_waterman_distance", "jaro_distance", "jaro_winkler_distance", "overlap_coefficient_distance",
          "generalized_jaccard_distance", "tfidf_distance", "soft_tfidf_distance", "partial_ration_distance", 
          "bag_distance_distance" ]] = minmax_scale(df[["levenshtain_distance", "needleman_wunsch_distance", 
                                                        "affine_gap_distance", "smith_waterman_distance", "jaro_distance", 
                                                        "jaro_winkler_distance", "overlap_coefficient_distance", 
                                                        "generalized_jaccard_distance", "tfidf_distance", "soft_tfidf_distance", 
                                                        "partial_ration_distance", "bag_distance_distance" ]], 
                                                        feature_range=(0, 1), axis=0)
    df.to_parquet(dest_path,engine="pyarrow")
    
```

```python
df_price_cpus = pd.read_parquet(str_path)
process_features(df_price_cpus,"../../Datos/features_parquet/df_price_cpus_select_12.parquet")
```
